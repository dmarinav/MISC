```python
def match_rows_priority(table_a, table_b, priority_keys, target_column):
    output_data = []
    added_rows = set()

    for a, row_a in table_a.iterrows():
        for b, row_b in table_b.iterrows():
            for priority in priority_keys:
                # Check if all values for the current priority are the same in both rows and are not 'all'
                if all(row_a[key] == row_b[key] and row_b[key] != 'all' for key in priority) and all(row_b[key] == 'all' for key in set(table_b.columns) - set(priority) - {target_column}):
                    key = tuple(row_a.values)
                    if key not in added_rows:
                        output_row = row_a.to_dict()
                        output_row.update({target_column: row_b[target_column]})
                        output_data.append(output_row)
                        added_rows.add(key)
                    
    _table = pd.DataFrame(output_data)
    return _table

# Example usage:
priority_keys = [('A', 'B', 'C'), ('A', 'B'), ('A', 'C'), ('B', 'C'), ('A',), ('B',), ('C',)]
target_column = 'target'
result_table = match_rows_priority(table_a, table_b, priority_keys, target_column)
print(result_table)
```
