```python
import pandas as pd

def merge_tables_priority(table_a, table_b, priority_keys):
    output_data = []
    added_rows = set()

    for a, row_a in table_a.iterrows():
        for b, row_b in table_b.iterrows():
            for keys in priority_keys:
                # The all() function checks if all items in a list are true.
                # The map() function checks each item in keys if it satisfies the conditions.
                # Here, the conditions are: key in both rows should be equal and not 'all'.
                if all(row_b[key] != 'all' for key in keys) and all(row_a[key] == row_b[key] for key in keys):
                    # Create a tuple to ensure the row is unique in the output data.
                    row_key = tuple(row_a[key] for key in table_a.columns)
                    if row_key not in added_rows:
                        output_data.append({
                            **row_a,
                            'target': row_b['target']
                        })
                        added_rows.add(row_key)
                    break

    return pd.DataFrame(output_data)


# Your tables
table_a = pd.DataFrame({
    'A': ['all', 1, 4, 4, 6, 5, 8, 'all'], 
    'B': [2, 2, 3, 2, 6, 7, 12, 'all'], 
    'C': [3, 3, 3, 3, 6, 1, 8, 13], 
    'D': [4, 4, 4, 4, 6, 1, 13, 9]
})

table_b = pd.DataFrame({
    'A': ['all', 1, 'all', 4, 5, 'all', 'all', 'all'], 
    'B': [2, 2, 2, 'all', 7, 'all', 'all', 'all'], 
    'C': [3, 3, 'all', 3, 'all', 8, 13, 'all'], 
    'target': ['xyz', 'zyx', 'tyz', 'abc', 'yyy', 'ppp', 'hgf', 'hhh']
})

# Priority keys
priority_keys = [('A', 'B', 'C'), ('A', 'B'), ('A', 'C'), ('B', 'C'), ('A',), ('B',), ('C',)]

# Call the function
result = merge_tables_priority(table_a, table_b, priority_keys)
print(result)
```
