```python
def res(resolution1,resolution2):
    if '/RES/' in resolution1:
        return resolution1
    if '/RES/' in resolution2 :
        return resolution2
    else:
        return 0

data['res']=data.apply(lambda x: res(x['res1'],x['res2'] ), axis=1)
```

```python
def res_split(res):
    for x in res.split(','):
        if '/RES/' in x:
            return x
data['symbol']=data['res'].apply(res_split)

```
